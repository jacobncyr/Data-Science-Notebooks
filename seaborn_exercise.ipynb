### Task 2: Plot a regression line between attack and defense while displaying a distinction between the different generations. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.lmplot(x="Attack", y="Defense", data=df, hue="Generation")
plt.show()
```

### Task 3: Make a boxplot for each feature of dataframe (without Generation and Legendary features) in one plot.

```python
import seaborn as sns

sns.catplot(data=df, kind="bar", x="Type 1", y="Type 2")

# Adding title and labels
plt.title("Bar Plot of Type 2 by Type 1")
plt.xlabel("Type 1")
plt.ylabel("Type 2")

# Displaying the plot
plt.show()


```

### Task 4: Plot a correlation heatmap of the pokemon dataframe (not including the features Generation and Legendary).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the Pokmon data from the provided URL
df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv', sep=',')
df.drop('#', axis=1, inplace=True)

# Drop the "Generation" and "Legendary" features
df = df.drop(["Generation", "Legendary"], axis=1)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create the correlation heatmap
plt.figure(figsize=(12, 8))  # Adjust the figure size if needed
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")

# Add a title
plt.title("Correlation Heatmap of Pokmon Features")

# Display the plot
plt.show()
```
