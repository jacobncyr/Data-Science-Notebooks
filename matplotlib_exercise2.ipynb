### Task 1: Read the `total_profit` of all months and display it using a line plot.

```python
x = df['month_number']
y = df['total_profit']

plt.plot(x,y)
```

### Task 2: Read all the different product sales data and display it using a multiline plot.

```python
x = df['facecream']
y = df['facewash']
z = df['toothpaste']
j = df['bathingsoap']
k = df['shampoo']
t = df['moisturizer']
plt.plot(x,y,z,j,k,t)
```

### Task 3: Read the `total_profit` of each month and display it using a histogram to find out which profit ranges are the most common.

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with the required columns

# Sort the DataFrame by 'month_number' to ensure the values are plotted in the correct order
df_sorted = df.sort_values('month_number')

# Extract the 'total_profit' column
y = df_sorted['total_profit']

# Create a list of month labels for the x-axis
x_labels = df_sorted['month_number']

# Plot the histogram
plt.bar(x_labels, y)

# Set the x-axis tick labels
plt.xticks(x_labels)

# Add labels and title to the plot
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.title('Total Profit by Month')

# Show the plot
plt.show()
```
